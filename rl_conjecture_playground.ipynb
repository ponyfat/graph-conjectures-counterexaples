```python
import torch
import networkx as nx
import matplotlib.pyplot as plt
```

```python
state = torch.load('run_7022_counterexample.pt')
```

```python
N = 19
# Construct the graph
G = nx.Graph()
G.add_nodes_from(list(range(N)))
count = 0
for i in range(N):
    for j in range(i + 1, N):
        if state[count] == 1:
            G.add_edge(i, j)
        count += 1
```

```python
nx.draw_kamada_kawai(G)
plt.show()
```
